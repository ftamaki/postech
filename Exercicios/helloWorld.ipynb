```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

# Gerar dados fictícios
np.random.seed(42)
categorias = ['A', 'B', 'C', 'D']
valores = np.random.randint(10, 100, size=40)
datas = pd.date_range('2023-01-01', periods=40, freq='D')
df = pd.DataFrame({
    'Data': datas,
    'Categoria': np.random.choice(categorias, size=40),
    'Valor': valores
})

# Estatísticas básicas
media_por_categoria = df.groupby('Categoria')['Valor'].mean()
maximo_por_categoria = df.groupby('Categoria')['Valor'].max()

# Plotar evolução dos valores por categoria
plt.figure(figsize=(10, 6))
for cat in categorias:
    subset = df[df['Categoria'] == cat]
    plt.plot(subset['Data'], subset['Valor'], label=f'Categoria {cat}')
plt.legend()
plt.title('Evolução dos Valores por Categoria')
plt.xlabel('Data')
plt.ylabel('Valor')
plt.grid(True)
plt.show()

# Exibir estatísticas
print("Média por categoria:")
print(media_por_categoria)
print("\nMáximo por categoria:")
print(maximo_por_categoria)
```
